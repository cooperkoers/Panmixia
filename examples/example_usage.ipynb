```python
import Panmixia as pmx
import geopandas as gpd
import os
```

# Naive Population Simulator

## Args
- N0: population size 0 (int)
- N1: population size 1 (int)
- L: genome length (int)
- mate_bias: mate bias weight for population size 1 (float)
- plotting: whether to plot figures (bool)

```python
# Example usage of the LocationPopulationSimulator
pmx.NaivePopulationSimulator(n0=200, 
                             n1=10, 
                             L=1000, 
                             mate_bias=1, 
                             plotting=True).simulate(generations=10)
```

```python
# print pwd
directory = "frames"
# if frame directory doesn't exist, create it
if not os.path.exists(directory):
    os.makedirs(directory)
for filename in os.listdir(directory):
    file_path = os.path.join(directory, filename)
    
    if os.path.isfile(file_path):
        os.remove(file_path)
```

```python
# load map data
map = gpd.read_file('data/maps/Climate_Zones_-_DOE_Building_America_Program.shp')
# select only texas and louisiana for faster plotting
map = map[map['BA_Climate'].isin(['Hot-Humid','Mixed-Humid', 'Hot-Dry'])]
```

# Spatial Population Simulator

## Args
- N0: population size 0 (int)
- N1: population size 1 (int)
- L: genome length (int)
- mate_bias: mate bias weight for population size 1 (float)
- plotting: whether to plot figures (bool)
- map: map object to simulate on (gpd object)

```python
pmx.LocationPopulationSimulator(n0=200, 
                                n1=10, 
                                L=1000, 
                                mate_bias=1, 
                                plotting=True, 
                                map=map).simulate(generations=10)
```
